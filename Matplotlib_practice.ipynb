```python
# import matpoltlib.pyplot as plt
from matplotlib import pyplot as plt
import numpy as np
x=np.array([1,2,4,5,6,7])
y=x*2
y1=3*x
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y,marker="*",c="r",s=100)
plt.scatter(x,y1,marker="o",c="g",s=100)
# plt.scatter(x,y2,marker="*",c="r",s=100)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker="*",c="r",s=100)
plt.title("x Vs Y")
plt.subplot(1,2,2)
plt.scatter(x,y1,marker="o",c="g",s=100)
plt.title("x Vs Y1")
plt.show()
```

```python
plt.hist(x,color="r")
plt.show()
```


# Box Plot

```python
one=[x for x in range(9)]
two=[j for j in range(3,12)]
three=[i for i in range(6,16)]
```

```python
data=list([one,two,three])
```

```python
plt.boxplot(data)
plt.show()
```

# Violin Plot

```python
plt.violinplot(data)
plt.show()
```

# pie -chart

```python
fruit=['apple','mangoes','guava','orange']
quantity=['10','20','22','7']
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','red','green','gray'])
plt.show()
```

# DoughNut Chart

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors='w',radius=1)
plt.show()
```
